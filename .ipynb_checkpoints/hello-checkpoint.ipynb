```python
import numpy as np
import math
import random

x = np.linspace(1, 20, 20)
y = []
print(y)

N = 100
r = 1

vol = 0

for D in range(20):
    vol = 0
    for i in range(N):
        L = 0
        for j in range(D):
            L += (random.uniform(-1, 1) * r)**2
        if L <= r**2:
            vol += 1

    #print(vol, vol/N * (2*r)**D, math.pi**(D/2) / math.gamma(D/2 + 1) * r**D)
    y.append(vol/N * (2*r)**D)
    
plt.plot(x, y, label="N = 100")

plt.xlabel("dimension")
plt.ylabel("volume")
plt.xticks( np.arange(0, 20, 2))
plt.legend()

plt.show()
```
